```python
df = pd.read_csv(r'DATA\power.csv')
df.head()
```

```python
df.info()
```

```python
countries = filter(lambda x: x.startswith('L'), df.country.unique())
for i in countries:
    print(i)
```

```python
# conditions = [
#     (df['Set'] == 'Z') & (df['Type'] == 'A'),
#     (df['Set'] == 'Z') & (df['Type'] == 'B'),
#     (df['Type'] == 'B')]
# choices = ['yellow', 'blue', 'purple']

countries = ['Estonia','Lithuania','Latvia']
categories = [4,12,21]
year_start = 2005
year_end = 2010



result = df[ ['country','quantity'] ][(df['country'].isin(countries)) & (df['quantity'] >= 0) & (df['category'].isin(categories)) & (df['year'].between(year_start,year_end))].groupby('country').agg({'quantity':'sum'})
result

```

### Задание №3
- [x] Выберите страницу любого сайта с табличными данными. Импортируйте таблицы в pandas dataframe.

```python
# url: https://en.wikipedia.org/wiki/Academy_Awards
import requests
from bs4 import BeautifulSoup


wikiurl="https://en.wikipedia.org/wiki/Academy_Awards"
table_class="wikitable sortable plainrowheaders jquery-tablesorter"
response=requests.get(wikiurl)
print(response.status_code)


# parse data from the html into a beautifulsoup object
soup = BeautifulSoup(response.text, 'html.parser')
oscars_ads = soup.find('table',{'class':"wikitable"})

df = pd.read_html(str(oscars_ads))
#  convert list to dataframe
df = pd.DataFrame(df[0])

df = df.rename(columns={"Viewers,millions[132]": "Viewers, mln","Ad price,[132][133]USD, millions": "Ad Price, mln USD", "Adjusted price,USD, millions":"Adjusted Price, mln USD"})
df
```
