## Задание 2. Работа с выбросами

В выбранных числовых столбцах найти выбросы, выдвинуть гипотезы об их причинах и проинтерпретировать результаты. Принять и обосновать решение о дальнейшей работе с ними.




```python
# проверяем типы данных для каждой колонки
dataTypeDict = dict(df.dtypes)
for key, val in dataTypeDict.items():
    print(f'{key} - {val}')
```

```python
# Убираем ошибки в категориальных данных

for col in df.columns:
    if df.dtypes[col] == 'object':
        print(f'{col} -- {df[col].unique()}')

        
df = df[df['2Age_bool'] != 9]
df = df[df['10Capillary_Refill_Time_bool'] != 3.0]
df
```

```python
# выбираем только числовые и для них считаем межквантильный размах, выбросы записываем в отдельный DF для изучения

outliers = pd.DataFrame()
    
for col in df.columns:
    # берем только float64 колонки
    if df.dtypes[col] == 'float64':
        q1 = df[col].quantile(0.25)
        q3 = df[col].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - (1.5 * iqr) 
        upper_bound = q3 + (1.5 * iqr) 
        print(lower_bound, upper_bound)
        # получаем DF без выбросов для этой колонки
        remove_outliers = df[df[col].between(lower_bound, upper_bound, inclusive=True)].dropna(subset=[col])
        # получаем строки с выбросами
        result = pd.concat([df.dropna(subset=[col]), remove_outliers]).drop_duplicates(keep=False)
        # добавляем значение колонки по какой колонке выбросы собираем
        result['VAR'] = col
        outliers = pd.concat([outliers, result])
outliers
```

Как описывали выше, уберем все случаи, когда пропущены измерения по температуре и пульсу, т.к. это базовые, дешевые измерения и их должны были делать для всех случаев диагностики. На основе этих параметров будем определять заполнения для недиагностированных показателей.

```python
df = df.dropna(subset=['4Rectal_Temp_C','5Pulse_bpm','1Surgery_bool'])
df.info()
```

```python
f = lambda x: x.median() if np.issubdtype(x.dtype, np.number) else next(iter(x.mode()), np.nan)
df['RT_rounded'] = df['4Rectal_Temp_C']
df['RT_rounded'] = df['RT_rounded'].apply(lambda x: round(x, 0))
df = df.fillna(df.groupby(['23Outcome_3cat','RT_rounded','5Pulse_bpm']).transform(f))                    
df.info()
```

```python
# Удалим пустые значения abdominal distension и повторим прошлую операцию при учете этого показателя.

df = df.dropna(subset=['13Abdominal_Distension_4cat'])
df = df.fillna(df.groupby(['13Abdominal_Distension_4cat','RT_rounded']).transform(f))  
df = df.fillna(df.groupby(['20Total_Protein_gms/dL']).transform(f))  
df.info()
```
